```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_stata('../data/external/ZA6647_v1-0-0.dta')
data.sample(10)
```

```python
cols_to_delete = ['studyno', 'doi', 'version', 'edition', 'survey', 'caseid', 'uniqid', 'country']
data = data.drop(columns=cols_to_delete)

cols_to_rename = {
    'd1': 'age',
    'd2': 'gender',
    'd5': 'job',
    'd13': 'town',
    'd22r': 'household size',
}
data = data.rename(columns=cols_to_rename)
data.sample(20)
```

```python
colnames = ['Country', 'Better and clearer information on the dates indicated on food labels', 'Better and clearer information on food product labels', 'Availability of smaller portion sizes in shops', 'Number of Votes']
lstq21 = [707, 336, 367, 354, 648, 521, 616, 462, 161, 298, 469, 516, 673, 475, 690, 363, 184, 417, 176, 398, 434, 183, 172, 641, 540, 263, 150, 289]
lstq22 = [668, 276, 332, 315, 653, 506, 655, 379, 144, 258, 376, 451, 680, 421, 655, 286, 155, 393, 124, 330, 398, 154, 142, 572, 561, 233, 116, 233]
lstq24 = [754, 609, 517, 203, 624, 635, 722, 547, 308, 318, 615, 705, 624, 485, 773, 430, 157, 399, 309, 485, 434, 231, 143, 578, 453, 291, 207, 251]
lstNumberOfVotes = [1001, 1007, 1000, 1000, 1007, 1001, 1008, 1000, 1003, 502, 1002, 1003, 1002, 1000, 1002, 1001, 503, 1000, 1000, 1003, 1003, 1000, 515, 1000, 1016, 1017, 1005, 1000]

lstcntry =['BELGIQUE', 'DANMARK', 'DEUTSCHLAND', 'ELLADA', 'ESPANA', 'SUOMI', 'FRANCE', 'IRELAND', 'ITALIA', 'LUXEMBOURG', 'NEDERLAND', 'ÖSTERREICH', 'PORTUGAL', 'SVERIGE', 'UK', 'BALGARIJA', 'KYPROS', 'CESKA REPUBLIKA', 'EESTI', 'MAGYARORSZAG', 'LATVIA', 'LIETUVA', 'MALTA', 'POLSKA', 'ROMANIA', 'SLOVENSKA REPUBLIC', 'SLOVENIJA', 'HRVATSKA']
df = pd.DataFrame(list(zip(lstcntry, lstq21, lstq22, lstq24, lstNumberOfVotes)), columns= colnames)
df.set_index('Country', inplace=True)
df['Better and clearer information on the dates indicated on food labels in percent'] = (df['Better and clearer information on the dates indicated on food labels'] / df['Number of Votes'] * 100).round(2)
df['Better and clearer information on food product labels in percent'] = (df['Better and clearer information on food product labels'] / df['Number of Votes'] *100).round(2)
df['Availability of smaller portion sizes in shops in percent'] = (df['Availability of smaller portion sizes in shops'] / df['Number of Votes'] * 100).round(2)
df
```

```python
df.iloc[:,[4,5,6]]
```

```python
avoid_waste_percent = df.iloc[:,[4,5,6]]
ax = avoid_waste_percent.plot.barh(title='What would help you to waste less food at home?')
ax.get_legend().remove()
fig = ax.get_figure()
fig.savefig('avoid waste in percent.png', dpi=300, facecolor='white', bbox_inches='tight')
```
